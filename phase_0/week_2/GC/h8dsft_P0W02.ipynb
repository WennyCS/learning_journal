## Image Preprocessing

### 1. Menghitung vektor gradien parsial

```python
grey = 0.299*Red + 0.587*Green + 0.114*Blue

dx, dy = np.gradient(grey)

fig, ax = plt.subplots(ncols=3,figsize=(16,5))
ax[0].imshow(grey)
ax[0].set_title('Real Image')
ax[1].imshow(dx)
ax[1].set_title(r'$\partial_x img$')
ax[2].imshow(dy)
ax[2].set_title(r'$\partial_y img$')
```

### 2. Menghitung gradient magnitude tiap pixel

```python
grad_max = np.sqrt((dx**2) +(dy**2))
grad_max
```

### 3. Membandingkan gradient magnitude dengan treshold lebih dari 30 dan 45

```python
fig,ax = plt.subplots(1,1, figsize=(16,9))
ax.imshow(grad_max >= 30, cmap='gray')
plt.axis('on')
```

```python
fig,ax = plt.subplots(1,1, figsize=(16,9))
ax.imshow(grad_max >= 45, cmap='gray')
plt.axis('on')
```

### 4. Lakukan operasi matriks sesuai aturan perkalian matriks

```python
plt.imshow(Red @ Blue)
plt.title('Gambar Red x Blue')
```

## Hasil dan Kesimpulan

Edge pada gambar semakin kecil ketika treshold gradient magnitute membesar. Gambar pada gradient magnitute dengan treshold di atas 45 mempunyai edge lebih kecil daripada gambar pada gradient magnitute dengan treshold di atas 30.
